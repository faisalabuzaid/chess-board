```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
class ChessBoard:
    x1=None
    y1=None
    x2=None
    y2=None
    
    def __init__(self):
        self.grid = np.zeros([8,8,3])
        for i in range(8):
            for j in range(8):
                if not ((i+j)%2):
                    self.grid[i,j] = [1,1,1]
    
          
    def render(self):
        """
        This method is to render Chess Board
        """
        plt.imshow(self.grid)

    def add_red(self,x,y):
        """
        This method is to add Red queen
        """
        self.grid[x,y]=[1,0,0]
        self.x1=x
        self.y1=y
        self.render()

    def add_blue(self,x,y):
        """
        This method is to add Blue queen
        """
        self.grid[x,y]=[0,0,1]
        self.x2=x
        self.y2=y
        self.render()
        
 
    def is_under_attack(self):
        """
        This method is to check if  blue and red queen will attack each other
        """
        if self.x1 == self.x2 or self.y1 == self.y2 or self.x2-self.x1 == self.y2-self.y1:
            return True
        return False
```

```python
chess=ChessBoard()
chess.render()
chess.add_red(1,1)
chess.add_blue(5,5)
assert chess.is_under_attack() == True #Diagonal check 
print('Diagonal Attack')
```

```python
chess=ChessBoard()
chess.render()
chess.add_red(1,7)
chess.add_blue(3,7)
assert chess.is_under_attack() == True #Vertical check
print('Vertical Attack')
```

```python
chess=ChessBoard()
chess.render()
chess.add_red(4,2)
chess.add_blue(4,5)
assert chess.is_under_attack() == True # Horizantal check
print('Horizantal Attack')
```

```python
chess=ChessBoard()
chess.render()
chess.add_red(1,2)
chess.add_blue(2,5)
assert chess.is_under_attack() == False #Not Under Attack
print('No Attack')
```
